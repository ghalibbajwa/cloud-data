```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
import chart_studio.tools as tls
```

```python
analytics = pd.read_csv('analytics.csv')
```

```python
analytics.columns=['id', 'session_id', 'avg_down', 'avg_jitter', 'avg_up', 'avg_rtt',
                           'max_down', 'max_jitter', 'max_up', 'max_rtt', 'min_down', 'min_jitter',
                           'min_rtt', 'min_up', 'r_packets', 'st_down', 'st_up', 'st_rtt',
                           't_packets', 'created_at', 'updated_at']
```

```python
meansdf=pd.DataFrame( columns=['session_id','mean_rtt','mean_up',"mean_down","mean_jitter",'through_put','totalsize','loss'])
```

```python
meansdf
```

```python
sessions=analytics['session_id'].unique()
```

```python
rtt=[]
down=[]
up=[]
jitter=[]
through_put =[]
totalsize = []
loss = []
for i in sessions:
    rtt.append(analytics[analytics['session_id'] == i]['avg_rtt'].mean())
    down.append(analytics[analytics['session_id'] == i]['avg_down'].mean())
    up.append(analytics[analytics['session_id'] == i]['avg_up'].mean())
    jitter.append(analytics[analytics['session_id'] == i]['avg_jitter'].mean())
    through_put.append((down[-1])+(up[-1]))
    totalsize.append((analytics[analytics['session_id'] == i]['r_packets'].sum())*50)
    loss.append((analytics[analytics['session_id'] == i]['t_packets'].sum())-(totalsize[-1]/50))
```

```python
meansdf['session_id'] = sessions
meansdf['mean_rtt'] = rtt
meansdf['mean_up'] = up
meansdf['mean_down'] = down
meansdf['mean_jitter'] = jitter
meansdf['through_put'] = through_put
meansdf['totalsize'] = totalsize
meansdf['loss']  = [abs(x) for x in loss]
```

```python
meansdf.to_csv('means.csv')
```
